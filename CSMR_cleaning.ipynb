```python
sheet_names = [ '1月','2月','3月','4月','5月','6月','7月','8月','9月','10月','11月','12月']
writer = pd.ExcelWriter('cleaned_data_dec.xlsx')
num = 1
for i in sheet_names:
    data = pd.read_excel('非微信.xlsx',sheet_name = i)
    data['网易链接'] = data['链接地址'].str.contains('dy\.163')
    data['网易链接2'] = data['链接地址'].str.contains('c\.m\.163')
    data['大风号'] = data['发布媒体'].str.contains('大风号')
    data['头条'] = data.作者.str.contains('今日头条')
    data = data.fillna({'头条': False})
    data_clean = data[
        (data.网易链接 == False) &
        (data.网易链接2 == False) &
        (data.大风号 == False) &
        (data.头条 == False)
    ]
    data_clean2 = data_clean.drop([
        '网易链接','网易链接2','大风号','头条'
    ],axis = 1)
    data_clean2.to_excel(excel_writer=writer, sheet_name=i, encoding="utf-8", index=False)
    print(i + "  保存成功！共%d个，第%d个。" % (len(sheet_names),num))
    num += 1
writer.save()
writer.close()
```
